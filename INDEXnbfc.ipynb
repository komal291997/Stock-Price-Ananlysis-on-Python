```python
#Plot based on opening price
bajaj['Open'].plot(label='Bajaj Finance', figsize=(5,5), title='Open Price')
muthoot['Open'].plot(label='Muthoot Finance')
sundaram['Open'].plot(label='Sundaram Finance')
plt.legend()
```

```python
#Plot based on closing price
bajaj['Adj Close'].plot(label='Bajaj Finance', figsize=(5,5), title='Close Price')
muthoot['Adj Close'].plot(label='Muthoot Finance')
sundaram['Adj Close'].plot(label='Sundaram Finance')
plt.legend()
```

```python
#Daily volume
bajaj['Volume'].plot(label='Bajaj Finance', figsize=(5,5), title='Volume')
muthoot['Volume'].plot(label='Muthoot Finance')
sundaram['Volume'].plot(label='Sundaram Finance')
plt.legend();
```

```python
#Plotting moving average of Bajaj
bajaj['MA15']=bajaj['Open'].rolling(15).mean()
bajaj['MA30']=bajaj['Open'].rolling(30).mean()
bajaj['MA100']=bajaj['Open'].rolling(100).mean()
bajaj[['Open', 'MA15', 'MA30', 'MA100']].plot(label='Bajaj Finance', figsize=(5,5))
```

```python
#Plotting moving average of Muthoot
muthoot['MA15']=muthoot['Open'].rolling(15).mean()
muthoot['MA30']=muthoot['Open'].rolling(30).mean()
muthoot['MA100']=muthoot['Open'].rolling(100).mean()
muthoot[['Open', 'MA15', 'MA30', 'MA100']].plot(label='Muthoot Finance', figsize=(5,5))
```

```python
#Plotting moving average of Sundaram
sundaram['MA15']=sundaram['Open'].rolling(15).mean()
sundaram['MA30']=sundaram['Open'].rolling(30).mean()
sundaram['MA100']=sundaram['Open'].rolling(100).mean()
sundaram[['Open', 'MA15', 'MA30', 'MA100']].plot(label='Sundaram Finance', figsize=(5,5))
```
